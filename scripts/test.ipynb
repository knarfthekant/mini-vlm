```python
from transformers import AutoTokenizer

# Test tokenizer
tokenizer = AutoTokenizer.from_pretrained("HuggingFaceTB/SmolLM2-360M-Instruct")
random_string = "hello"
templated = tokenizer.apply_chat_template([{"role": "assistant", "content": random_string}], tokenize=False,
            add_generation_prompt=False)
tokenized = tokenizer.apply_chat_template([{"role": "assistant", "content": random_string}], tokenize=True,
            add_generation_prompt=False,
            return_dict=True,
            truncation=True)
print(tokenizer.pad_token_id)
print
print(templated)
print(tokenized)
```
